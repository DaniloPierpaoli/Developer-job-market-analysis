# Data Visualization

Here we try to gain further insights from data that are not part of our main focus (see Trend Charts notebook).

```python
#Import the relevant libraries 

import sqlite3
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

mpl.style.use('ggplot')  

# Load dataset
df = pd.read_csv("survey-data_normalized.csv", keep_default_na= False)

# Connect to the SQLite database
conn = sqlite3.connect('survey-data_norm.sqlite')

# Write the dataframe to the SQLite database
df.to_sql('main', conn, if_exists='replace', index=False)
```

### Visualizing yearly compensation distribution by age range

```python

QUERY = """
SELECT Age, ConvertedCompYearly
FROM main 
WHERE Age = '18-24 years old'
"""

df_18to24 = pd.read_sql_query(QUERY, conn)

QUERY = """
SELECT Age, ConvertedCompYearly
FROM main 
WHERE Age = '25-34 years old'
"""

df_25to34 = pd.read_sql_query(QUERY, conn)

QUERY = """
SELECT Age, ConvertedCompYearly
FROM main 
WHERE Age = '35-44 years old'
"""

df_35to44 = pd.read_sql_query(QUERY, conn)

QUERY = """
SELECT Age, ConvertedCompYearly
FROM main 
WHERE Age = '45-54 years old'
"""

df_45to54 = pd.read_sql_query(QUERY, conn)

#Create 4 subplots
fig = plt.figure(figsize=(12, 12))

# Add the first subplot (top-left)
ax1 = fig.add_subplot(2, 2, 1)
ax1.hist(df_18to24["ConvertedCompYearly"], bins = 15)
plt.xlim(0,350000)
ax1.set_title('18 to 24 years old group')
ax1.set_xlabel('Yearly Compensation / $')
# Add the second subplot (top-right)

ax2 = fig.add_subplot(2, 2, 2)
ax2.hist(df_25to34["ConvertedCompYearly"], bins = 15)
plt.xlim(0,350000)
ax2.set_title('25 to 34 years old group')
ax2.set_xlabel('Yearly Compensation / $')

# Add the third subplot (bottom-left)
ax3 = fig.add_subplot(2, 2, 3)
ax3.hist(df_35to44["ConvertedCompYearly"], bins = 15)
plt.xlim(0,350000)
ax3.set_title('35 to 44 years old group') 
ax3.set_xlabel('Yearly Compensation / $')

# Add the fourth subplot (bottom-right)
ax4 = fig.add_subplot(2, 2, 4)
ax4.hist(df_45to54["ConvertedCompYearly"], bins = 15)
plt.xlim(0,350000)
ax4.set_title('45 to 54 years old group')
ax4.set_xlabel('Yearly Compensation / $')
#Adding a Title for the Overall Figure
fig.suptitle('Yearly compensation distribution by age range', fontsize=15)

plt.show()
```

We can clearly see that the lower the age range the more the distribution is skewed to the left. Older people are paid more.

### Compensation developer vs non developer

```python
QUERY = """
SELECT MainBranch, ConvertedCompYearly AS YearlyComp
FROM main
"""
main_branch = pd.read_sql_query(QUERY, conn)

branch_map = {
    "I am a developer by profession": "Developer" }



# Map the categories
main_branch["MainBranch"] = main_branch["MainBranch"].map(branch_map).fillna("Not a developer")




main_branch = main_branch.groupby("MainBranch")["YearlyComp"].mean()

main_branch.plot(kind = "barh", x = "MainBranch", y = "YearlyComp", figsize = (10,6), color = "blue")
plt.xlabel("Yearly Compensation / $")
plt.title("Chart showing Yearly compensation for developer and not developers as main profession")
plt.show()
```

The mean value for the yearly compensation seem to be slightly higher for developers than non developers.

### Coding experience distribution

```python
QUERY = """
SELECT YearsCodePro
FROM main 
WHERE YearsCodePro != 'None'
"""
df_yearsExp = pd.read_sql_query(QUERY, conn)
df_yearsExp.replace(to_replace= ["Less than 1 year","More than 50 years"], value=[0, 55], inplace= True)
df_yearsExp = df_yearsExp.dropna()
df_yearsExp = df_yearsExp.astype(int)
median = df_yearsExp["YearsCodePro"].median()
df_yearsExp.plot(kind = "hist", figsize = (12,7),legend = False, bins = range (0,60,5), color = "orange")
plt.title("Years of experience coding among respondents")
plt.xlabel("Years")
plt.axvline(median, color = "red", label = f"Median: {median} years")
plt.legend()
plt.show()
```
